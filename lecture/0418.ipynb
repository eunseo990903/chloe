```python
import pandas as pd
```

```python
suni=pd.read_excel('./data/서울지역대학교위치.xlsx')
suni.head()
```

```python
#지도에 표시 
!pip install folium
```

```python
import folium
seoulmap=folium.Map(location=[37.55,126.98],zoom_start=12)
seoulmap
```

```python
for name,lat,lng in zip(suni.대학교,suni.위도, suni.경도):
    folium.Marker(location=[lat,lng],tooltip= name).add_to(seoulmap)
    #marker 마킹, 위도경도 지도에 마킹해주고 서울맵에 추가 
seoulmap
```

```python
seoulmap.save('seouluniversity.html')
```

```python
for name,lat,lng in zip(suni.대학교,suni.위도, suni.경도):
    folium.CircleMarker(location=[lat,lng],radius=10, color= 'white',fill='True',
                        fill_color='red', fill_opacity=0.6,tooltip= name).add_to(seoulmap)
    #marker 마킹, 위도경도 지도에 마킹해주고 서울맵에 추가 
seoulmap
```
