```python
df = pd.DataFrame([[100, 'C100', '컴퓨터구조', 91],
                     [200, 'C200', '데이터베이스', 85],
                     [400, 'C300', '운영체제', 65],
                     [200, 'C300', '운영체제', 85],
                     [300, 'C100', '컴퓨터구조', 100],
                     [500, 'C200', '데이터베이스', 55]],
                     index = range(1,7), columns=['학번','과목번호','과목이름','점수'])
df 

#데이터프레임 중첩이면 가로로, 딕셔너리로는 세로 
```

```python
#등급 for문, 삼항 if 문 
grade = []
for i in df['점수']:
    if i >= 90: grade.append("A")
    elif i >= 80: grade.append("B")
    elif i <= 70: grade.append("C")
df['등급'] = grade 



df['등급']= ["A" if i >= 90 else "B" if i >= 80 else 'C' for i in df['점수'] ]
df
```

```python
학번 200 인 사람들의 총점과 평균 
df[df['학번']==200]['점수'].sum()
df[df['학번']==200]['점수'].mean()
```

```python
#A인 사람 몇명 isin, sum(), .__len__() 
df['등급'].isin(['A']).sum()
(df['등급'] == "A").sum()
df[df['등급']=="a"].__len__() 
```

```python
for i,gen in enumerate(movies.genres):
    indices=dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i,indices]=1
```
