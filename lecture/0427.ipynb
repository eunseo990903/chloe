```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from numpy import nan as NA
import pydataset
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
from pydataset import data
from sklearn.preprocessing import LabelEncoder

mpg= data('mpg')
encoder = LabelEncoder()
encoder.fit(mpg['class'])
mpg['class']= encoder.transform(mpg['class']) 

mpg[:1]

x_train,x_te,y_train,y_te= train_test_split(mpg.iloc[:,[2,4,7,8]], mpg['class'] ,test_size=0.2)
x_tr,x_val,y_tr,y_val=train_test_split(x_train,y_train,test_size=0.1) 
#validation train 나눌때 train 안헷갈리게 tr, train 으로 나눔 

evals=[(x_tr,y_tr),(x_val,y_val)] #바로 위 걸로 묶음 , early stopping 을 위한 검증세트 
from xgboost import XGBClassifier
xgb_clf= XGBClassifier(n_estimators= 400 ,learning_rate=0.05, max_depth=3)

xgb_clf.fit(x_tr,y_tr,early_stopping_rounds=50, eval_metric='mlogloss', eval_set = evals)

pred=xgb_clf.predict(x_te)
accuracy_score(y_te,pred).round(4)
print(f"{accuracy_score(y_te,pred):.4}")
```

```python
from hyperopt import hp
#search space: x,y 와 범위  -10~ 10까지 1간격 갖는 x, -15~15 까지 1간격 갖는 y
search_space={'x': hp.quniform('x',-10,10,1), 'y':hp.quniform('y',-15,15,1)}

from hyperopt import STATUS_OK
def objective_func(search_space):
    x=search_space['x']
    y=search_space['y']
    retval=x**2-20*y
    return retval  #반환은 결과값으로 x**2-2y는 안나오고 뭘 최소로 넣어야 성립할지 찾아내는것 
    
```

```python
from hyperopt import fmin,tpe,Trials
#입력결과값저장 Trials 객체값 생성
trial_val= Trials()

#5번의 입력값 시도 
best1=fmin(fn=objective_func,space=search_space, algo=tpe.suggest, max_evals=5, 
           trials=trial_val, rstate=np.random.default_rng(seed=0))
print('best',best1) # x=-4, y=12  (x^2-20y의 최적은 x=0, y=15 최소값 )
#20번 
best2=fmin(fn=objective_func,space=search_space, algo=tpe.suggest, max_evals=20, 
           trials=trial_val, rstate=np.random.default_rng(seed=0))
print('best',best2)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from hyperopt import STATUS_OK
dataset = load_breast_cancer()

cancer_df = pd.DataFrame( data=dataset.data, columns=dataset.feature_names)
cancer_df['target']= dataset.target

X_features = cancer_df.iloc[:, :-1] #target이 보통 맨뒤에 있어서 -1 
y_label = cancer_df.iloc[:, -1]

# 전체 데이터 중 80%는 학습용 데이터, 20%는 테스트용 데이터 추출
X_train, X_test, y_train, y_test=train_test_split(X_features, y_label, test_size=0.2, random_state=156 ) 
X_tr,X_val,y_tr,t_val=train_test_split(X_train,y_train, test_size=0.1, random_state=156)
```

```python
from hyperopt import hp
#search_space 만들기 
xgb_clf=XGBClassifier()
xgb_search_space={'max_depth':hp.quniform('max_depth',5,20,1), #5~20까지 1씩 증가 
                 'min_child_weight':hp.quniform('min_child_weight',1,2,1),
                 'learning_rate': hp.uniform('learning_rate', 0.01, 0.2), #uniform의 범위 정해주기 
                 'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1),
                   } 
```

```python
#목적함수
from sklearn.model_selection import cross_val_score
def objective_func(search_space):
    xgb_clf=XGBClassifier(n_estimatiors=100, max_depth=int(search_space['max_depth']),
                         min_child_weight=int(search_space['learning_rate']),
                         learning_rate=search_space['colsample_bytree'],
                         eval_metric='logloss')
    accuracy = cross_val_score(xgb_clf,X_train,y_train,scoring='accuracy', cv=3)
    
    #accuracy는 cv=3 만큼 roc - auc 결과 리스트 가짐. 평균해서 반환하되 -1 곱함
    return {'loss': -1*np.mean(accuracy),'status':STATUS_OK}
```

```python
from hyperopt import fmin,tpe,Trials

trial_val=Trials()
best=fmin(fn=objective_func,space=xgb_search_space, algo=tpe.suggest, max_evals=50, #최대반복횟수 
           trials=trial_val, rstate=np.random.default_rng(seed=9))
import warnings
warnings.filterwarnings('ignore')
```

```python
print('best:',best)
```
