```python
#movielens
pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('./data/movielens/users.dat', sep='::',
                      header=None, names=unames, engine='python',encoding='UTF-8')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('./data/movielens/ratings.dat', sep='::',
                        header=None, names=rnames, engine='python',encoding_errors='ignore')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('./data/movielens/movies.dat', sep='::',
                       header=None, names=mnames, engine='python',encoding_errors='ignore') 

#merge 3개 안됨.
movie= pd.merge(pd.merge(users,ratings,on ='user_id'),movies, on='movie_id')
movie[:3]

```

```python
#나이가 20-40 이며 여자인 데이터만 추출, 가장 많은 평점 개수가 있는 여성 상위 10개 추출
#query , 콤마 안되고 and, &
movie.query(' gender=="F" & 20<=age & age<=40').groupby('title').size().sort_values(ascending=False)[:10]

#pivot 사용하여 성별에 따른 각 영화의 평균 
movie.pivot_table('rating', columns= 'gender', index='title',aggfunc='mean')
#agg 안써도 변화없음 
#남성에게 높은 평점을 받은 영화목록 5개 
movie.pivot_table('rating', columns= 'gender', index='title',aggfunc='mean').sort_values(by="M", ascending=False)[:5]
#남녀의 평점 합을 구해 movie_ sum 컬럼 만들기
a=movie.pivot_table('rating', columns= 'gender', index='title',aggfunc='mean')
a['movie_sum']=a.sum(axis=1)
a
```

```python
def foreign_data(yy, mm):
# 1. 불러올 엑셀 파일 경로 지정
    file_path = './data/foreign/foreign_{}{}.xlsx'.format(yy, mm)
# 2. 엑셀 파일 불러오기
    im = pd.read_excel(file_path, header = 1, skipfooter = 4) #skipfooter 아래에서 4개 행 없애기 
# 3. '기준년월' 컬럼 추가
    im['기준년월'] = '{}-{}'.format(yy, mm)

    return im

# 6자리로 정렬하여 기준년월 출력하기 
for yy in range(2010, 202):
    for mm in range(1, 13):
        mm_str = str(mm).zfill(2)   
# zfill을 적용하여 10월 이전의 달들도 01, 02,
        yymm = '{}{}'.format(yy, mm)
    
im = pd.DataFrame()

for yy in range(2010, 2021):
    for mm in range(1, 13):
        if (yy == 2020 and mm > 5):
            break
        tmp = foreign_data(str(yy), str(mm).zfill(2))
        im = im.append(tmp, ignore_index = True)
im['국적'].unique()
```
