```python
import pandas as pd

# 데이터를 저장할 리스트
data = []

driver = webdriver.Chrome()
url='http://bank.weatherimc.co.kr/webstats/calendar/month.php?obs_id=108&yy=2023&mm=1'
driver.get(url)
#지역 
for a in range (1,80):
    elements = driver.find_element(By.XPATH, '//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[1]').click()
    driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[1]/option[{a}]').click()
    reigion =driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[1]/option[{a}]').text


#년도 
    for r in range (20,24):
        elements = driver.find_element(By.XPATH, '//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[2]').click()
        driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[2]/option[{r}]').click()
        year =driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[2]/option[{r}]').text

# 월
        for z in range(1, 13):
            elements = driver.find_element(By.XPATH, '//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[3]').click()
            driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[3]/option[{z}]').click()
            month =driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[1]/tbody/tr[2]/td[1]/select[3]/option[{z}]').text

    # 일
            try:
                for i in range(2, 8):
                    for j in range(1, 8):
                        day = driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[2]/tbody/tr[{i}]/td[{j}]/div/font/b').text
                        weather = driver.find_element(By.XPATH, f'//*[@id="contentWrap"]/div[2]/form/table[2]/tbody/tr[{i}]/td[{j}]/font/p').text
                        data.append([reigion,year, month, day, weather])
            except Exception:
                pass

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['Reigion','Year','Month','Day', 'Weather'])

#\n으로 새로운 열 만들기
df[['Condition','Min Temp','Max Temp','Avg Temp','Rainfall' ]]= df['Weather'].str.split('\n', expand=True)

df.drop(['Weather', 'Condition'], axis=1, inplace=True)
#' :' 앞에 문자열 삭제 
for column in df.iloc[:,[4,5,6,7]]:
    df[column] = [value.split(':')[-1].strip() if isinstance(value, str) else value for value in df[column]]
df
```

```python
df['Month']=df['Month'].str.replace('월','')
df
```

```python
#숫자형으로 변환해서 정렬 
df['Day'] = pd.to_numeric(df['Day'], errors='coerce')
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df=df.sort_values(['Reigion','Year', 'Month', 'Day'])
df
```

```python
# 중복된 Year, Month, Day를 가진 행 중에서 앞에 있는 행을 제거
df =df.drop_duplicates(subset=['Reigion','Year', 'Month', 'Day'], keep='last')
df
```

```python
import calendar

# 연도별 월별 일수 계산 함수
def get_days_in_month(year, month):
    days_in_month = calendar.monthrange(year, month)[1]
    return days_in_month

# 각 연도의 월별 일수 계산
df['DaysInMonth'] = df.apply(lambda row: get_days_in_month(row['Year'], row['Month']), axis=1)

# 일수를 고려하여 행 제거
df_cleaned = df[df['Day'] <= df['DaysInMonth']]

# 불필요한 칼럼 삭제
df_cleaned = df_cleaned.drop('DaysInMonth', axis=1)
```
