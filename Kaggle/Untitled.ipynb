```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
model.compile(loss='mean_squared_error', optimizer='adam', run_eagerly=True)

# 데이터 전처리
features = ['평균기온','평균최고기온','평균최저기온','평균습도','월합강수량',
            '평균풍속','최대풍속','최심적설','평균지면온도'	]  # 예측에 사용할 특성 선택
data =sweather[features]

# 데이터 정규화
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# 학습 데이터 생성
look_back = 36  # 과거 36개 달의 데이터를 기반으로 다음 달의 날씨를 예측
X = []
y = []
for i in range(len(data_scaled) - look_back):
    X.append(data_scaled[i:i+look_back])
    y.append(data_scaled[i+look_back])
X = np.array(X)
y = np.array(y)

# 학습 데이터와 검증 데이터 분할
train_size = int(len(X) * 0.8)
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]

# LSTM 모델 구축
model = Sequential()
model.add(LSTM(64, input_shape=(look_back, len(features))))
model.add(Dense(len(features)))
model.compile(loss='mean_squared_error', optimizer='adam')

# 모델 학습
model.fit(X_train, y_train, epochs=20, batch_size=500, validation_data=(X_val, y_val))

# 2023년 날씨 예측
last_months = data_scaled[-look_back:]  # 최근 36개 달의 데이터
prediction = []
for _ in range(12):  # 12개 달에 대한 예측
    pred = model.predict(np.array([last_months]))
    prediction.append(pred)
    last_months = np.concatenate((last_months[1:], pred), axis=0)
prediction = np.array(prediction)
prediction = scaler.inverse_transform(prediction)  # 정규화된 예측값을 원래 값으로 변환

# 예측 결과 출력
prediction_df = pd.DataFrame(prediction, columns=features)
prediction_df['년'] = 2023
prediction_df['월'] = range(1, 13)
print(prediction_df)
```
