```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor


# 피처와 타겟 변수 선택
features = ['평균기온(°C)', '평균최고기온(°C)', '평균최저기온(°C)', '평균상대습도(%)', '월합강수량(00~24h만)(mm)', '평균풍속(m/s)', '합계 일조시간(hr)', '최심적설(cm)', '평균지면온도(°C)']
target_date = '일시'
target_region = 'eocs지역명'

# 피처 데이터와 타겟 변수 분할
X = pa_weather[features]
y_date = pa_weather[target_date]
y_region = pa_weather[target_region]

# 레이블 인코딩을 위한 전처리기
label_encoder = LabelEncoder()

# 날짜 타겟 변수 레이블 인코딩
y_date_encoded = label_encoder.fit_transform(y_date)

# 지역 타겟 변수 레이블 인코딩
y_region_encoded = label_encoder.fit_transform(y_region)

# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_date_train, y_date_test, y_region_train, y_region_test = train_test_split(X, y_date_encoded, y_region_encoded, test_size=0.2, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

# 데이터 로드
df = pd.read_csv('weather_eocs.csv', encoding='cp949')


# '일시' 컬럼 데이터 타입을 datetime으로 변환
df['일시','eocs지역명'] = pd.to_datetime(df['일시'])

# 입력 변수와 타겟 변수로 분할
X = df[['일시']]
y = df[['eocs지역명', '평균기온(°C)', '평균최고기온(°C)', '평균최저기온(°C)', '평균상대습도(%)', '최소상대습도(%)', '월합강수량(00~24h만)(mm)']]

# 'eocs지역명'을 원-핫 인코딩
eocs_encoded = pd.get_dummies(y['eocs지역명'], prefix='eocs')
y = pd.concat([eocs_encoded, y.drop('eocs지역명', axis=1)], axis=1)

# 훈련 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 다중 선형 회귀 모델 초기화
model = MultiOutputRegressor(RandomForestRegressor(random_state=42))

# 모델 학습
model.fit(X_train, y_train)

# 예측
predictions = model.predict(X_test)

# 예측 결과 출력
print(predictions)
```
